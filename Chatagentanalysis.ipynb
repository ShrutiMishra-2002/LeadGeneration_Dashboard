#Relative Metrics

```python

df['Conversion (%)'] = (df['total_leads'] / df['total_chats']) * 100
df['RPC'] = df['total_revenue'] / df['total_chats']
df['Greeting Conversion (%)'] = (df['total_chats'] / df['total_greetings_fireds']) * 100
df['Engagement (%)'] = (df['total_chats'] / df['total_traffics']) * 100
```

Function for calculating top_n and bottom_n for each metrics.

```python
def get_top_bottom_n(df, metrics, n=50):
    results = {}

    df_cleaned = df.drop_duplicates(subset=['client_id'])

    for metric in metrics:
        if metric in df_cleaned.columns:
            top_n = df_cleaned.nlargest(n, metric)
            bottom_n = df_cleaned.nsmallest(n, metric)
            results[metric] = {
                'top_n': top_n,
                'bottom_n': bottom_n
            }
        else:
            print(f"Metric '{metric}' does not exist in the DataFrame.")

    return results

metrics = ['revenue', 'total_chat', 'Conversion (%)', 'RPC', 'Engagement (%)']
top_bottom_results = get_top_bottom_n(df, metrics, n=50)
```

#Trend Analysis (last 6 months)

```python
df['date_'] = pd.to_datetime(df['date_'])
start_date = '2024-03-01'
end_date = '2024-09-30'

last_six_months = df[(df['date_'] >= start_date) & (df['date_'] <= end_date)]

last_six_months.head()
```

```python
last_six_months['total_revenue'] = pd.to_numeric(last_six_months['total_revenue'], errors='coerce')

last_six_months['Month'] = pd.to_datetime(last_six_months['date_']).dt.to_period('M').astype(str)

top_clients = last_six_months[['client_name', 'total_revenue']].groupby('client_name').sum()['total_revenue'].nlargest(5)
bottom_clients = last_six_months[['client_name', 'total_revenue']].groupby('client_name').sum()['total_revenue'].nsmallest(5)

print("Last 6 months trend for Top Clients based on Total Revenue")
top_clients
```

```python
top_clients.plot(
    kind='bar', x='client_name',y='total_revenue', color='green',figsize=(8, 5),title='Top 5 Clients based on Total Revenue in past 6 months'
)
plt.xticks(rotation=45)
```

```python
print("\nLast 6 months trend for Bottom Clients based on Total Revenue:")
bottom_clients
```

```python
bottom_clients.plot(
    kind='bar', x='client_name', y='total_revenue', color='orange',figsize=(8, 5),title='Bottom 5 Clients based on Total Revenue in past 6 months'
)
plt.ylim(200000, 290000)
plt.yticks(range(200000, 290000, 15000))
plt.xticks(rotation=45)
```

Client buckets are built based on the full months revenue (except august)

```python
df['Month'] = pd.to_datetime(df['date_']).dt.to_period('M')

# Filtering out August 2024 as it's incomplete
filtered_data = df[df['Month'] != '2024-08']

# calculating total revenue for each client in every month
revenue_per_client_month = (
    filtered_data.groupby(['client_name'])['revenue']
    .sum()
)
filtered_data[['client_name', 'revenue']].head(10)
```

```python
revenue_per_client_month.plot(kind='hist', bins=10, title='Client Revenue Distribution')
plt.xlabel('Monthly Revenue')
plt.show()
```

```python
revenue_per_client_month.head(5)
```

```python
def bucket_revenue(revenue):
    if revenue < 500:
        return 'Bronze'
    elif revenue < 1500:
        return 'Silver'
    elif revenue < 3000:
        return 'Gold'
    else:
        return 'Platinum'

revenue_buckets = revenue_per_client_month.apply(bucket_revenue)
print("Client's segregation based on revenue bucket")
revenue_buckets.head(5)
```

#Clients that generate a lot of traffic but aren't engaging much in terms of chats.

```python
 # High traffic clients above the 75 percentile
high_traffic_threshold = df['total_traffic'].quantile(0.75)
 # Low engagement clients below the 25 percentile
low_engagement_threshold = df['Engagement (%)'].quantile(0.25)

# Filtering clients with High Traffic but Low Engagement
high_traffic_low_engagement = df[(df['total_traffic'] > high_traffic_threshold) &
                                  (df['Engagement (%)'] < low_engagement_threshold)]

high_traffic_low_engagement = high_traffic_low_engagement[['client_name', 'total_traffic', 'total_chats', 'Engagement (%)', 'RPC']]

print("Clients with high traffic and low engagement")
high_traffic_low_engagement.head()
```

Idea : These clients needs to be targetted for increasing their engagement by managing traffic in correct ways

#Top performing months

```python
# Aggregating by months
monthly_data = df.groupby('month_').agg(
    total_revenue=('total_revenue', 'sum'),
).reset_index()

monthly_data.plot(
    x='month_',y=['total_revenue'],figsize=(8, 4),marker='o',title='Performance Across Months',xlabel='Month',ylabel='Values',grid=True
)

plt.show()
```
